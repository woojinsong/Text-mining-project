```python
import pandas as pd
import csv
import jpype
from konlpy.tag import Okt
```

```python
a=pd.read_csv('C:/Users/wooji/Desktop/새 폴더 (3)/lexicon/polarity.csv')
table=dict()
with open('polarity.csv','r',-1,'utf-8') as polarity:
    next(polarity)
    
    for line in csv.reader(polarity):
        key =str()
        for word in line[0].split(';'):
            key +=word.split('/')[0]
        table[key]={"부정":line[3],'중립':line[4],'긍정':line[6]}
```

```python
title=["hankyoreh","경향신문","동아일보","머니투데이","서울경제","조선일보","중앙일보","프레시안"]
subject=["박영선","오세훈","대통령","백신","이낙연","이재명"]
```

```python
neg=0
net=0
pos=0
result1=[]
result2=[]
result3=[]
for x in range(0,6):
    for y in range(0,8):
        data=pd.read_csv(subject[x]+"_"+title[y]+".csv",encoding='utf-8-sig')
        splitter=Okt()
        for j in data["content"]:
            neg=0
            net=0
            pos=0
            k=splitter.morphs(j)
            for i in k:
                try:
                    neg=neg+float(table[i]["부정"])
                    net+=float(table[i]["중립"])
                    pos+=float(table[i]["긍정"])
                except:
                    continue
            print("긍정:",pos)
            print("부정:",neg)
            print("중립:",net)
            result1.append(pos)
            result2.append(neg)
            result3.append(net)
        data["긍정"]=result1
        data["부정"]=result2
        data["중립"]=result3
        data.to_csv(subject[x]+"_"+title[y]+"_result.csv",encoding='utf-8-sig')
```
